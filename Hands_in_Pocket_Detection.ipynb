```python
import cv2
import mediapipe as mp
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Initialize MediaPipe Pose
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(static_image_mode=True, min_detection_confidence=0.5)
mp_drawing = mp.solutions.drawing_utils

# Upload files
uploaded = files.upload()

# Process each uploaded file
for filename in uploaded.keys():
    print(f"\nProcessing uploaded image: {filename}")
    # Read the image
    image = cv2.imdecode(np.frombuffer(uploaded[filename], np.uint8), cv2.IMREAD_COLOR)
    if image is None:
        print(f"Error: Could not load image from {filename}")
        continue

    # Convert the image to RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    results = pose.process(image_rgb)

    if results.pose_landmarks:
        print("Pose landmarks detected.")

        # Draw landmarks
        annotated_image = image.copy()
        mp_drawing.draw_landmarks(
            annotated_image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
            mp_drawing.DrawingSpec(color=(245, 117, 66), thickness=2, circle_radius=2),
            mp_drawing.DrawingSpec(color=(245, 66, 230), thickness=2, circle_radius=2)
        )

        # Show image
        plt.imshow(cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.title(f"Processed Image: {filename}")
        plt.show()

        # Extract landmarks
        lm = results.pose_landmarks.landmark
        left_wrist, right_wrist = lm[mp_pose.PoseLandmark.LEFT_WRIST], lm[mp_pose.PoseLandmark.RIGHT_WRIST]
        left_hip, right_hip     = lm[mp_pose.PoseLandmark.LEFT_HIP], lm[mp_pose.PoseLandmark.RIGHT_HIP]
        left_shoulder, right_shoulder = lm[mp_pose.PoseLandmark.LEFT_SHOULDER], lm[mp_pose.PoseLandmark.RIGHT_SHOULDER]

        # Shoulder width
        shoulder_width = abs(left_shoulder.x - right_shoulder.x)

        # Thresholds
        pocket_threshold_x = 0.30 * shoulder_width
        pocket_threshold_y = 0.45 * shoulder_width

        left_hand_in_pocket = (
            abs(left_wrist.x - left_hip.x) < pocket_threshold_x and
            (left_wrist.y >= left_hip.y - pocket_threshold_y) and
            (left_wrist.y <= left_hip.y + pocket_threshold_y)
        )

        right_hand_in_pocket = (
            abs(right_wrist.x - right_hip.x) < pocket_threshold_x and
            (right_wrist.y >= right_hip.y - pocket_threshold_y) and
            (right_wrist.y <= right_hip.y + pocket_threshold_y)
        )


        if left_hand_in_pocket or right_hand_in_pocket:
            print("Possible hands in pockets detected.")
            if left_hand_in_pocket:
                print(" - Left hand might be in pocket.")
            if right_hand_in_pocket:
                print(" - Right hand might be in pocket.")
        else:
            print("No hands in pockets detected.")
    else:
        print("No pose landmarks detected.")

pose.close()
```
